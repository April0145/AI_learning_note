特征缩放

1，Z-score标准化

```python
# 手动Z-score标准化
import pandas as pd

# 计算均值和标准差
mean = titanic['Age'].mean()
std = titanic['Age'].std()

tn_age = titanic[['Age']]
# 手动标准化
Standardized_Age = (tn_age - mean) / std
# 将标准化后的数据 Standardized_Age 转换为 DataFrame，并保持原列名称。
df_scaled = pd.DataFrame(Standardized_Age, columns=tn_age.columns)

titanic = titanic.drop('Age', axis=1)
titanic = pd.concat([titanic, df_scaled], axis=1)
titanic.head()
```

2，均值标准化

```python
# 手动实现均值标准化
import pandas as pd

# 计算 'Age' 列的均值、最小值和最大值
mean_age = titanic['Age'].mean()
min_age = titanic['Age'].min()
max_age = titanic['Age'].max()

# 提取 'Age' 列
tn_age = titanic[['Age']]

# 手动均值归一化 'Age' 列
Mean_Normalized_Age = (tn_age - mean_age) / (max_age - min_age)

# 创建均值归一化后的 DataFrame
df_mean_normalized = pd.DataFrame(Mean_Normalized_Age, columns=tn_age.columns)

# 更新原数据集
titanic = titanic.drop('Age', axis=1)
titanic = pd.concat([titanic, df_mean_normalized], axis=1)
titanic.head()
```
